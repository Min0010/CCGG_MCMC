```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
os.listdir()
os.listdir('CC_dataset/')
```

```python
#CC data
full_data = pd.read_excel('CC_dataset/CC_data.xlsx')
full_data
```

```python
redshift = full_data['redshift']
Hub = full_data['H(z)']
Hub_unc = full_data['sigma']
```

```python
plt.figure()
plt.errorbar(redshift, Hub, yerr=Hub_unc, capsize=3, fmt='r.', label='CC data')
plt.xlabel('$z$')
plt.ylabel('$H\,\,\,[km/s/Mpc]$')
plt.legend()
plt.show()
```
